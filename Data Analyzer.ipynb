# Data Analyzer

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def data_reader():
    try:
        df = pd.read_json('data.json')
    except ValueError:
        print('Incorrect file name')
    return df
```

```python
DATA = data_reader()
```

```python
DATA.head()
```

```python
def coffee_description():
    #Inspect the Dataframe
    print(DATA.describe())
    
coffee_description()
```

```python
def acidity():
    #Visualize coffee acidity distribution
    df_sort_date = DATA.sort_values("Fecha")
    acidity = df_sort_date["Acidez"]
    plt.hist(acidity, bins=10)
    plt.xlabel('Acidity')
    plt.show() 

acidity()
```

```python
DATA.groupby(['fragancia_cualitativa']).mean().to_frame().hist(figsize=(20,20))
```
